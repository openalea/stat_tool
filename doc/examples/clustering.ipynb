This notebook illustates how to perform clustering with stat_tool.Cluster

```python
from openalea.stat_tool import get_shared_data
from openalea.stat_tool.data_transform import SelectVariable
from openalea.stat_tool.comparison import Compare
from openalea.stat_tool.cluster import Clustering

from openalea.stat_tool.vectors import VectorDistance, Vectors

# from pathlib import Path

# import pytest

# 
vec10 = Vectors(get_shared_data("chene_sessile.vec"))
```

```python
from openalea.stat_tool.data_transform import SelectVariable
vec15 = SelectVariable(vec10, [1, 3, 6], Mode="Reject")
```

```python
# computation of a distance matrix using a standardization procedure

matrix10 = Compare(vec15, VectorDistance("N", "N", "N"))
```

```python
# clustering using a partitioning method

Clustering(matrix10, "Partition", 2).display()
```

```python
Clustering(matrix10, "Hierarchy", "Agglomerative")
```

```python
Clustering(matrix10, "Hierarchy", "Divisive")
```
